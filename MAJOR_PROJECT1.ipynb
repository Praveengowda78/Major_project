```python
#4.Divide data into input and output
#Input (x) is always 2 dimentional , output(y) is 1 dimensional
#Input(x)-AREA COLUMN, Output(y)-Prices column
#In column's place, if there is a ':' then array is  2 dimensional
#In colum's palce , if there is no ":", then array is 1 dimensional
```

```python
x=df.iloc[:,0:1].values #  .values gives 1 dimensional
x
```

```python
y=df['Salary'].values
y
```

```python
#5. Train and test variables- If we have more data, preform step no. 5 better accuracy
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
# the variables names can change, but the variable order cannot change
```

```python
#After train_test_split
#75% of data in x_train ,rest goes to x_test
#75% of data in y_train ,rest goes to y_test
```

```python
#7. Run a classifier, regressor, cluster(applying a suitable algorith)
#sklearn.linear_model- package(collection of libraries) ,LogisticRegression-library name
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
#8. Fit the model(mapping or plotting the inputs with the output)
#LinearRegression.fit(x,y)
model.fit(x,y) #We are plotting the values of x and y inside the LinearRegression library
```

```python
#9. Predict the output
y_pred=model.predict(x) # using the input values we predict the output
y_pred
```

```python
# conclusion: We have to comapre y_pred and y.
#So when we compare , we come to know there is a huge difference.
#This huge diference does not mean that my modl has predicted wrong.
# It only means that my model is not linear/less linear
#Linearity of a moel depemds on the nature of the data and size of the data
```

```python
y_test #Actual output
```

```python
plt.scatter(x,y) #Actual output values
plt.plot(x_test,y_pred,c='orange') #Predicted output values
plt.title('Best fit line')
plt.xlabel('Area')
plt.ylabel("Prices")
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
